```python
# Dataset Introduction

The Boston Housing dataset contains information about house prices 
and different factors affecting them such as crime rate, number of rooms, 
tax rate, pollution levels, and population.
The purpose of this project is to analyze the housing dataset and
understand which factors influence house prices. 
This project uses Exploratory Data Analysis (EDA) and visualizations 
to generate meaningful insights from the data.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv"
df = pd.read_csv(url)
df.head()
```

```python
df = df.drop_duplicates()
```

```python
#histogram
plt.figure(figsize=(6,4))
plt.hist(df["medv"], bins=20)
plt.title("House Price Distribution")
plt.xlabel("House Price")
plt.ylabel("Count")
plt.show()
```

```python
#scatterplot
plt.figure(figsize=(6,4))
plt.scatter(df["rm"], df["medv"])
plt.title("Rooms vs House Price")
plt.xlabel("Average Rooms")
plt.ylabel("House Price")
plt.show()
```

```python
#boxplot
plt.figure(figsize=(6,4))
sns.boxplot(y=df["crim"])
plt.title("Crime Rate Distribution")
plt.show()
```

```python
#correlation
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
columns = ["rm", "tax", "ptratio"]
avg = df[columns].mean()
plt.figure(figsize=(6,4))
avg.plot(kind='bar')
plt.title("Average Values of Selected Features")
plt.show()
```

```python
# Insights and Findings

1. Houses with more rooms generally have higher prices.

2. Areas with high crime rates tend to have lower housing prices.

3. Tax rate and neighborhood conditions affect house prices.

4. Most house prices fall in the medium price range.

5. Some variables strongly influence house value.
```

```python
# Final Conclusion / Recommendations

Based on the analysis, house prices are affected by several factors such as number of rooms, crime rate, and neighborhood conditions.

People planning to buy houses should prefer areas with lower crime and better facilities for long-term value.
```
